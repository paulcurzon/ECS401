### Solution to Exercise 3<a id="Solution3"></a>
This is a much better program than the original because of the use of methods to break it up into distinct parts, each with a clear purpose.

```java
// A simple chatbot that says hello, asks questions and then says goodbye

//    Ask for a string with given message
//    Return the string typed in by the user 
//
    public static String inputString(String message)
    {
       Scanner scanner = new Scanner(System.in);
       String answer;

       System.out.println(message);
       answer = scanner.nextLine();
   
       return answer;
    } // END inputString

// Print a greeting (a hello message)
//
public static void helloMessage ()
{
    System.out.println("Hello. My name is Chatty. I'm a chatbot.");
    return;
} // END helloMessage


// Ask a person's name
//
public static String askQuestion1name ()
{
    String name;
     
    name = inputString("What’s your name?");
    return name;
}


// Ask how a person is using their name
//
public static void askQuestion2howareyou (String name)
{
    String howareyou;
   
    howareyou = inputString("Hi there " + name + " How are you?");
    System.out.println("I'm glad to hear it " + name + ".");
    return;
}

// Ask a person's subject
//
public static void askQuestion3subject ()
{
    String subject;
     
    subject = inputString("What subject are you doing?");
    System.out.println("I nearly took " + subject + " but in the end I went for Linguistics.");
    return;
}

// Ask a person's favourite music
//
public static void askQuestion4music ()
{
    String music;
     
    music = inputString("What type of music do you like?");
    System.out.println("I like " + music + " music too.");
    return;
}

// Ask if computers worry them
//
public static String askQuestion5worry ()
{
    String worry;
     
    worry = inputString("Do computers worry you?");
    return worry;
}

// Ask why computers are a worry
//
public static void askQuestion6whyWorry (String worry)
{
    String reason;
    
    reason = inputString("Why do you say " + worry + "?");
    System.out.println("Really? I think that too!");
}
// Ask questions and respond to the answers
//
public static void askQuestions ()
{
    String name;
    String worry;
    
    name = askQuestion1name();
    askQuestion2howareyou(name);
    askQuestion3subject();
    askQuestion4music();
    worry = askQuestion5worry();
    askQuestion6whyWorry(worry);

    return;
} // END askQuestions

// Print a salutation (a goodbye message)
//
public static void goodbyeMessage ()
{
    System.out.println("Oops. Thats my phone, I have to go now.  I've enjoyed talking with you.");
    return;
} // END goodbyeMessage

// A friendly chatbot that greets, asks a question and says goodbye
//
public static void mychatbot3 ()
{
    helloMessage();
    askQuestions();
    goodbyeMessage();
    return;
} // END hellomessage

mychatbot3();
```
