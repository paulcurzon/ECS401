## What is an array
We often need to store and manipulate bulk data of the same type, such as the names of all the players in a tennis tournament; or all the marks of students from a test. That is what arrays are for.

Arrays are a way to store lots of values, instead of just one, in a single variable. All the values MUST be **of the same type**. You can think of a variable of type array as being storage space that has lots of separate compartments, like a cabinet full of lots of identically sized drawers, with one thing stored in each drawer. 

A program can get at all values using the single variable name given to the array. If the array is called, *names*, then each name is accessed using the label *names*. The separate compartments are referred to by a number which gives the position or **index** of where it can be found. So, it is as though all the drawers are lined up in order and each drawer of the cabinet has a number labelling it. To get at a particular drawer, to get a copy out or put something new in, you give the cabinet's name and the number of the drawer.


The index position of an array starts at **0**, so **the first position in an array is the 0th position**.

## Summary so far
## Declaring and initialising an array

The array type is just the type of the values that can be stored in that array, followed by square brackets []. So to **declare** an array of Strings called level we would write:
```java
String[] level;
```
We can **initialise** an array with a value when it is declared by putting the list of values for each **element** in curly brackets. So to store String values LOW, MEDIUM and HIGH in an array called *level* when it is declared we would write
```java
String[] level = {"LOW", "MEDIUM", "HIGH"};
```

## Accessing the value in an array

To **access** a value in an array we just give the **index number**, remembering that they start counting at 0. So to access the value "MEDIUM"  in the above array we would write.
```java
level[1]
```
If we wanted to print that value we would write
```java
System.out.println(level[1]);
```
If we wanted to get a copy of the value out and store it, on its own, in a String variable called current_level, we would write
```java
String current_level = level[1];
```
You can read this as **"Variable *current_level* GETS A COPY OF THE VALUE stored in array *level* at position 1"**

## Storing values in an array
You store values in an array in a similar way to the way you get copies of values out, by giving the **array name** and the **index position**. Suppose we have created an array called *level* using

````java
String[] level = {"LOW", "MEDIUM", "HIGH"};
````

To **overwrite** the old value with "MIDDLE" in  position 1 of an array called *level* we would write

````java
level[1] = "MIDDLE";
````

We can use a variable to hold the position of the array we want to access (whether putting a value in or taking a value out). So the following does the same as the above and then prints the value.

````java
int position = 1;
level[position] = "MIDDLE";
System.out.println(level[position]);
````

It is this ability to store the index in a variable that gives arrays great flexibility over just having a different variable for each value you want to store.

### Solution to Exercise 12<a id="Solution12"></a> (Summary so far)

Arrays are a way to store lots of data of the same type, so lots of equivalent information about different things like the salaries of lots different people, the home town of lots of different people, or all the scores in a video game for a high score table.

Array values are, therefore, compound values. They combine lots of simpler values of the same type (eg all ints or all Strings) into one new compound value that can be stored in a single variable with a single name. 

The separate values stored in the array are accessed by their **index**. The index is just the storage position: a number counting from 0. You can use the index to refer to the separate positions of an array variable. So for an array called *a*, ```a[0]``` is position 0 (the first position), ```a[1]``` is position 1, and so on. You can think of these as similar to the names of individual variables. You can print values stored in an array or assign new values in particular positions in this way. For example,
```java
a[5] = "Hello";
```
stores a new value in position 5 of array *a*, and 
```java
System.out.println(a[5]) prints that value.
```

An array is more flexible than having lots of separate variables as we can replace the index by a variable that the program changes. As we will see in the next notebook, that allows us to process arrays using for loops.
